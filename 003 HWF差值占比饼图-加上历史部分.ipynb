```python
# alldata = pd.read_csv('Reclassify_mean_2041_2100_HWF-difference.csv')
alldata = pd.read_excel('Reclassify_mean_HWF-difference.xlsx', sheet_name=0)
print(alldata)
```

```python
colors =  [ '#4575B4', '#FFEBAF', '#FF0000', '#FFD37F', '#FFAA00']

#   浅蓝色 #ABD9E9  深蓝色 #4575B4  淡黄色 #FFEBAF    橙色#FFAA00  深红色#FF0000   浅黄色 #FFD37F
```

```python
data = alldata["245_historical"]

# plt.rcParams["font.weight"] = "bold"
fig, axe = plt.subplots(figsize=(1, 1), dpi=300)
# 调整图片的边距
# plt.subplots_adjust(left=0.025, right=0.945, top=0.99, bottom=0.0009)
wedges, texts, autotexts= plt.pie(data,colors = colors,pctdistance = 1.2,autopct='%.0f%%')
plt.setp(texts, size=10, weight='bold', color='black')

plt.savefig('01 245_historical.png', dpi=300,facecolor="none", transparent=True)
plt.show()
```
